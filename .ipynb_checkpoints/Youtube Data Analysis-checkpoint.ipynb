## Objective:
- Let's collect the data about the trending videos on youtube using <b>"YouTube Data API v3"</b> to analyze and find what makes a video trend in youtube.

```python
# Importing necessary libraries
import pandas as pd
from googleapiclient.discovery import build
import warnings
warnings.filterwarnings("ignore")

# API Key
API_KEY = "AIzaSyAF1ezeJAsWTVQAovXeXsQjdutqjs0vmx4"
region = "IN"

def get_video_categories(api_key, region_code = region):
    # Build YouTube service
    youtube = build('youtube', 'v3', developerKey = api_key)
    
    # Fetching the video categories for the given region
    request = youtube.videoCategories().list(
        part = 'snippet',
        regionCode = region_code
    )
    response = request.execute()
    
    # Creating a dictionary to map categoryId to categoryName
    categories = {item['id']: item['snippet']['title'] for item in response['items']}
    
    return categories

def get_trending_videos(api_key, max_results = 200, region_code = region):
    
    # Build YouTube service
    youtube = build('youtube', 'v3', developerKey = api_key)
    
    # List to hold video details
    videos = []
    
    # Fetch video categories
    categories = get_video_categories(api_key, region_code)
    
    # Fetching the most popular videos
    request = youtube.videos().list(
        part = 'snippet, contentDetails, statistics',
        chart = 'mostPopular',
        regionCode = region_code,
        maxResults = 50
    )
    
    # Paginate through the results if max_results > 50
    while request and len(videos) < max_results:
        response = request.execute()
        for item in response['items']:
            category_id = item['snippet']['categoryId']
            video_details = {
                'video_id': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'published_at': item['snippet']['publishedAt'],
                'channel_id': item['snippet']['channelId'],
                'channel_title': item['snippet']['channelTitle'],
                'category_id': category_id,
                'category_name': categories.get(category_id, 'Unknown'),
                'tags': item['snippet'].get('tags', []),
                'duration': item['contentDetails']['duration'],
                'definition': item['contentDetails']['definition'],
                'caption': item['contentDetails'].get('caption', 'false'),
                'view_count': item['statistics'].get('viewCount', 0),
                'like_count': item['statistics'].get('likeCount', 0),
                'dislike_count': item['statistics'].get('dislikeCount', 0),
                'favorite_count': item['statistics'].get('favoriteCount', 0),
                'comment_count': item['statistics'].get('commentCount', 0)    
            }
            videos.append(video_details)

        # Get the next page token
        request = youtube.videos().list_next(request, response)

    return videos[:max_results]

# Saving collected data
def save_to_csv(data, fileName):
    df = pd.DataFrame(data)
    df.to_csv(fileName, index=False)

def main():
    trending_videos = get_trending_videos(API_KEY)
    fileName = 'trending_videos.csv'
    save_to_csv(trending_videos, fileName)
    print(f'Trending videos saved to {fileName}')

if __name__ == '__main__':
    main()

```

```python
# Grouping to get average of view, like and comments count

average_matrix_category_name = trending_videos.groupby('category_name')[['view_count','like_count','comment_count']].mean().sort_values(by = 'view_count',
                                                                                                                          ascending = False)
average_matrix_category_name['category_type'] = average_matrix_category_name.index.map(category_mapping)
average_matrix_category_name
```

```python
average_matrix_category_type = trending_videos.groupby('category_type')[['view_count',
                                                                         'like_count','comment_count']].mean().sort_values(by = 'view_count',
                                                                                                                          ascending = False)
average_matrix_category_type
```

```python
# Function to plot graph based on category name

# Plot by category name
def count_plot_by_category_name(y, x, palette, ax):
    sns.barplot(y=y, x=x, data=average_matrix_category_name.reset_index(), ax=ax, hue='category_type', palette=palette)
    ax.set_title(f"Average {x} by Category Name")
    ax.set_xlabel(f"Average {x}")
    if ax == axes[0]:
        ax.set_label("Category Name")
    else:
        ax.set_ylabel("")
        ax.set_yticks([])

# plot by category type
def count_plot_by_category_type(y, x, palette, ax):
    sns.barplot(y=y, x=x, data=average_matrix_category_type.reset_index(), ax=ax,  palette=palette)
    ax.set_title(f"Average {x} by Category Type")
    ax.set_xlabel(f"Average {x}")
    if ax == axes[0]:
        ax.set_label("Category Type")
    else:
        ax.set_ylabel("")
        ax.set_yticks([])
```

```python
# Setting up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 4))

# Define color palette for the bars
palette = {'Entertainment': '#0033CC', 'Life style': '#3399FF', 'Professional': '#99CCFF'}

# Plotting the averages
count_plot_by_category_name(y='category_name', x='view_count', palette=palette, ax=axes[0])
count_plot_by_category_name(y='category_name', x='like_count', palette=palette, ax=axes[1])
count_plot_by_category_name(y='category_name', x='comment_count', palette=palette, ax=axes[2])
# Remove legends from other axes
for ax in axes[0:]:
    ax.legend_.remove()
# Setting up the figure and axes
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(18, 2))


# Plotting the averages
count_plot_by_category_type(y='category_type', x='view_count', palette=palette, ax=axes[0])
count_plot_by_category_type(y='category_type', x='like_count', palette=palette, ax=axes[1])
count_plot_by_category_type(y='category_type', x='comment_count', palette=palette, ax=axes[2])

plt.tight_layout()
plt.show()
```

```python
# Convertiof of duration for ISO 8601 to 

import isodate 

# from ISO 8601 format to total seconds
trending_videos['duration_seconds'] = trending_videos['duration'].apply(lambda x: isodate.parse_duration(x).total_seconds())

# binnig duration_seconds into 5 categories 
bin_labels = ['0-1 min','1-5 min', '5-10 min', '10-20 min', '20-60 min', '60-120 min', '>120 min']
trending_videos['duration_category'] = pd.cut(trending_videos['duration_seconds'], 
                                              bins =[0, 60, 300, 600, 1200, 3600, 7200, float('inf')],
                                             labels = bin_labels,
                                             right = False)
trending_videos['duration_category'].value_counts().reindex(bin_labels)
```

# <u><center><b>CONCLUSION</b></center></u>


## 1. Distribution Skewness
- The analysis reveals a **right-skewed distribution** across views, likes, and comments, indicating a limited number of videos with high engagement metrics, while the majority have lower counts.

## 2. Correlation Insights
- A **strong positive correlation** exists between:
  - **View Count** and **Like Count**: Videos with higher views tend to accumulate more likes.
  - **Like Count** and **Comment Count**: An increase in likes is often accompanied by an increase in comments.

## 3. Category Distribution
- Approximately **75% of trending videos** are categorized under **Entertainment**.
- About **9% of the videos** pertain to **Professional Growth**, including themes such as Education and Politics.

## 4. Engagement Trends
- **Entertainment videos** demonstrate consistent growth in both likes and comments as view counts increase.
- **Lifestyle videos** exhibit a higher average comment count relative to likes.
- **Music videos** attract substantial views but tend to receive fewer average likes and comments.
- **People & Blogs** categories show lower average views, yet their likes and comments are significantly higher.
- **Science & Technology** videos consistently garner both likes and comments as views rise.

## 5. Video Duration
- Videos with a duration of **1-5 minutes** achieve optimal engagement in terms of views, likes, and comments.
- **Short videos** (0-1 minute) generally receive fewer comments in relation to their views and likes.
- Videos lasting **20-60 minutes** generate higher view and like counts but receive comparatively fewer comments.

## 6. Tags and Views
- The analysis indicates no strong relationship between the number of tags and views; however, using **10-25 tags** is recommended for enhanced visibility.
- A significant number of videos utilize approximately **20-30 tags**.

## 7. Optimal Posting Times
- Consistent posting is observed between **8 AM and 4 PM**.
- Videos posted between **7 AM to 9 AM** and **3 PM to 5 PM** tend to achieve higher initial view counts.
- Posting during **6 PM to 8 PM** is associated with slower view growth.
- The **optimal time to post** is identified as between **7 AM to 5 PM**, while the **least effective time** is **6 PM to 8 PM**.
